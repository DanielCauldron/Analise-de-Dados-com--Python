Importe as bibliotecas: Para realizar a análise de dados, precisa importar as bibliotecas necessárias. As bibliotecas mais comuns são pandas, numpy e matplotlib.digite os seguintes comandos:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv("Car_sales.csv") #carregar os dados que você quer analisar. 
```

```python
dados.head() # Exibe as primeiras linhas dos dados
dados.info() # Exibe informações sobre as colunas dos dados
dados.describe() # Exibe estatísticas descritivas dos dados
```

```python
dados.dropna() # Remove linhas com valores ausentes
dados.drop_duplicates() # Remove linhas duplicadas
dados.replace() # Substitui valores antigos por novos
```

```python
dados.mean() # Calcula a média das colunas
dados.corr() # Calcula a correlação entre as colunas
plt.plot(dados['__year_resale_value']) # Cria um gráfico de linha para a coluna
```

```python
plt.hist(dados['Sales_in_thousands']) # Cria um histograma para a coluna
```
